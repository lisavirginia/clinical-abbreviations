```python
df = pd.read_csv('https://raw.githubusercontent.com/lisavirginia/clinical-abbreviations/master/code/data/Errors_Annotated.csv',
                 sep='|',
                 header=0,
                 index_col=False,
                 na_filter=False,
                 dtype=object)
```

```python
df = df.rename(columns={"EntryID": "RecordID"})
```

```python
df['RecordID'] = df['RecordID'].str.replace('E','R')
```
